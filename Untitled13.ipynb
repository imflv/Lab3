```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
print("="*60)

df = pd.read_excel('gym_membership.xlsx')

print("Data loaded successfully!")
print("="*60)

print("\nFirst 5 rows of data:")
print("="*40)
print(df.head())
print("="*40)

print(f"\nDataset shape: {df.shape}")
print(f"Number of members: {df.shape[0]:,}")
print(f"Number of columns: {df.shape[1]}")

print("\nData types:")
print("="*40)
print(df.info())

print("\nStatistical summary:")
print("="*40)
print(df.describe())

print("\nColumn names:")
print("="*40)
for i, col in enumerate(df.columns, 1):
    print(f"{i:3}. {col}")

print("\nMissing values:")
print("="*40)
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Percentage': missing_pct.round(2)
})
print(missing_df[missing_df['Missing Count'] > 0])

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap', fontsize=16)
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

print("\nHandling missing values...")
df_clean = df.copy()

df_clean['fav_group_lesson'].fillna('None', inplace=True)
df_clean['fav_drink'].fillna('No Drink', inplace=True)
df_clean['name_personal_trainer'].fillna('No Trainer', inplace=True)

print("Missing values handled!")

print("\nAfter cleaning - Missing values:")
print(df_clean.isnull().sum())

print("\nChecking for duplicates...")
duplicates = df_clean.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    df_clean.drop_duplicates(inplace=True)
    print(f"Removed {duplicates} duplicate rows")
else:
    print("No duplicates found!")

print("\nCreating new features...")

# Fixed line - convert to string first
df_clean['check_in_hour'] = pd.to_datetime(df_clean['avg_time_check_in'].astype(str)).dt.hour

def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df_clean['visit_time'] = df_clean['check_in_hour'].apply(time_of_day)

bins = [0, 18, 25, 35, 45, 55, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior', 'Elder']
df_clean['age_group'] = pd.cut(df_clean['Age'], bins=bins, labels=labels)

df_clean['visit_category'] = pd.cut(df_clean['visit_per_week'],
                                    bins=[0, 1, 2, 3, 5],
                                    labels=['Light', 'Regular', 'Frequent', 'Very Frequent'])

df_clean['total_activities'] = (df_clean['attend_group_lesson'].astype(int) +
                                df_clean['personal_training'].astype(int) +
                                df_clean['drink_abo'].astype(int))

print("New features added!")
print(f"New columns: check_in_hour, visit_time, age_group, visit_category, total_activities")

print("\nCreating visualizations...")

fig = plt.figure(figsize=(20, 16))

ax1 = plt.subplot(3, 3, 1)
gender_counts = df_clean['gender'].value_counts()
colors = ['#ff6b6b', '#4ecdc4']
ax1.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90)
ax1.set_title('Gender Distribution', fontsize=14)

ax2 = plt.subplot(3, 3, 2)
membership_counts = df_clean['abonoment_type'].value_counts()
ax2.bar(membership_counts.index, membership_counts.values, color=['#f9ca24', '#f0932b'])
ax2.set_title('Membership Type', fontsize=14)
ax2.set_xlabel('Type')
ax2.set_ylabel('Count')

ax3 = plt.subplot(3, 3, 3)
ax3.hist(df_clean['Age'], bins=30, edgecolor='black', alpha=0.7, color='#3498db')
ax3.axvline(df_clean['Age'].mean(), color='red', linestyle='--',
            label=f'Mean: {df_clean["Age"].mean():.1f}')
ax3.set_title('Age Distribution', fontsize=14)
ax3.set_xlabel('Age')
ax3.set_ylabel('Frequency')
ax3.legend()

ax4 = plt.subplot(3, 3, 4)
visit_counts = df_clean['visit_per_week'].value_counts().sort_index()
ax4.bar(visit_counts.index, visit_counts.values, color='#6c5ce7')
ax4.set_title('Visits per Week', fontsize=14)
ax4.set_xlabel('Number of Visits')
ax4.set_ylabel('Count')

ax5 = plt.subplot(3, 3, 5)
lesson_counts = df_clean['attend_group_lesson'].value_counts()
ax5.pie(lesson_counts.values, labels=['Attend', 'Not Attend'], autopct='%1.1f%%',
        colors=['#00b894', '#dfe6e9'], startangle=90)
ax5.set_title('Group Lesson Attendance', fontsize=14)

ax6 = plt.subplot(3, 3, 6)
drink_counts = df_clean['drink_abo'].value_counts()
ax6.bar(['No Drink', 'Have Drink'], drink_counts.values, color=['#ff7675', '#74b9ff'])
ax6.set_title('Drink Subscription', fontsize=14)
ax6.set_ylabel('Count')

ax7 = plt.subplot(3, 3, 7)
trainer_counts = df_clean['personal_training'].value_counts()
ax7.pie(trainer_counts.values, labels=['No Trainer', 'Have Trainer'], autopct='%1.1f%%',
        colors=['#a29bfe', '#fd79a8'], startangle=90)
ax7.set_title('Personal Training', fontsize=14)

ax8 = plt.subplot(3, 3, 8)
sauna_counts = df_clean['uses_sauna'].value_counts()
ax8.bar(['No Sauna', 'Uses Sauna'], sauna_counts.values, color=['#fdcb6e', '#e17055'])
ax8.set_title('Sauna Usage', fontsize=14)
ax8.set_ylabel('Count')

ax9 = plt.subplot(3, 3, 9)
time_counts = df_clean['visit_time'].value_counts()
ax9.bar(time_counts.index, time_counts.values, color='#f39c12')
ax9.set_title('Preferred Visit Time', fontsize=14)
ax9.set_xlabel('Time of Day')
ax9.set_ylabel('Count')
ax9.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\nCreating advanced visualizations...")

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.scatterplot(data=df_clean, x='Age', y='visit_per_week', hue='gender', alpha=0.6)
plt.title('Age vs Visits per Week by Gender', fontsize=14)

plt.subplot(2, 2, 2)
membership_gender = pd.crosstab(df_clean['abonoment_type'], df_clean['gender'])
membership_gender.plot(kind='bar', ax=plt.gca())
plt.title('Membership Type by Gender', fontsize=14)
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.subplot(2, 2, 3)
numeric_cols = ['Age', 'visit_per_week', 'avg_time_in_gym', 'check_in_hour', 'total_activities']
corr_matrix = df_clean[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix', fontsize=14)

plt.subplot(2, 2, 4)
sns.boxplot(data=df_clean, x='abonoment_type', y='avg_time_in_gym', hue='gender')
plt.title('Time in Gym by Membership Type and Gender', fontsize=14)
plt.xlabel('Membership Type')
plt.ylabel('Time (minutes)')

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("PATTERNS AND INSIGHTS")
print("="*60)

print("\nINSIGHT 1: Gender Analysis")
print("-"*40)
gender_stats = df_clean.groupby('gender').agg({
    'visit_per_week': 'mean',
    'avg_time_in_gym': 'mean',
    'attend_group_lesson': 'mean',
    'personal_training': 'mean'
}).round(2)

gender_stats.columns = ['Avg Visits', 'Avg Time', 'Group %', 'Personal %']
gender_stats = gender_stats * 100
print(gender_stats)

print("\nINSIGHT 2: Premium vs Standard")
print("-"*40)
membership_stats = df_clean.groupby('abonoment_type').agg({
    'visit_per_week': 'mean',
    'avg_time_in_gym': 'mean',
    'attend_group_lesson': 'mean',
    'uses_sauna': 'mean'
}).round(2)

membership_stats.columns = ['Visits', 'Time', 'Group %', 'Sauna %']
membership_stats = membership_stats * 100
print(membership_stats)

print("\nINSIGHT 3: Age Groups")
print("-"*40)
age_stats = df_clean.groupby('age_group', observed=True).agg({
    'visit_per_week': 'mean',
    'avg_time_in_gym': 'mean',
    'attend_group_lesson': 'mean'
}).round(2)

age_stats.columns = ['Visits', 'Time', 'Group %']
age_stats = age_stats * 100
print(age_stats)

print("\nINSIGHT 4: Peak Hours")
print("-"*40)
time_dist = df_clean['visit_time'].value_counts()
for time, count in time_dist.items():
    pct = (count/len(df_clean))*100
    print(f"{time}: {count} members ({pct:.1f}%)")

print("\nINSIGHT 5: Top Trainers")
print("-"*40)
trainer_counts = df_clean[df_clean['name_personal_trainer'] != 'No Trainer']['name_personal_trainer'].value_counts()
for trainer, count in trainer_counts.head().items():
    print(f"{trainer}: {count} clients")

print("\nINSIGHT 6: Favorite Drinks")
print("-"*40)
all_drinks = []
for drinks in df_clean[df_clean['fav_drink'] != 'No Drink']['fav_drink']:
    all_drinks.extend([d.strip() for d in str(drinks).split(',')])

drink_counts = pd.Series(all_drinks).value_counts()
for drink, count in drink_counts.head().items():
    print(f"{drink}: {count} orders")

print("\nINSIGHT 7: Favorite Group Lessons")
print("-"*40)
all_lessons = []
for lessons in df_clean[df_clean['fav_group_lesson'] != 'None']['fav_group_lesson']:
    all_lessons.extend([l.strip() for l in str(lessons).split(',')])

lesson_counts = pd.Series(all_lessons).value_counts()
for lesson, count in lesson_counts.head().items():
    print(f"{lesson}: {count} attendees")

print("\n" + "="*60)
print("FINAL SUMMARY - KEY FINDINGS")
print("="*60)

print(f"""
MEMBERSHIP OVERVIEW:
   • Total Members: {len(df_clean):,}
   • Male: {df_clean['gender'].value_counts()['Male']/len(df_clean)*100:.1f}%
   • Female: {df_clean['gender'].value_counts()['Female']/len(df_clean)*100:.1f}%
   • Premium Members: {df_clean['abonoment_type'].value_counts()['Premium']/len(df_clean)*100:.1f}%

BEHAVIOR PATTERNS:
   • Average Visits/Week: {df_clean['visit_per_week'].mean():.2f}
   • Average Time in Gym: {df_clean['avg_time_in_gym'].mean():.0f} minutes
   • Peak Time: {df_clean['visit_time'].mode()[0]}
   • Group Lesson Attendance: {df_clean['attend_group_lesson'].mean()*100:.1f}%
   • Sauna Usage: {df_clean['uses_sauna'].mean()*100:.1f}%

TOP FINDINGS:
   1. Women visit {abs(df_clean[df_clean['gender']=='Female']['visit_per_week'].mean() - df_clean[df_clean['gender']=='Male']['visit_per_week'].mean()):.2f} more times per week than men
   2. Premium members are {membership_stats.loc['Premium', 'Visits'] - membership_stats.loc['Standard', 'Visits']:.1f}% more active
   3. Most active age group: {age_stats['Visits'].idxmax()}
   4. Most popular trainer: {trainer_counts.index[0] if len(trainer_counts)>0 else 'None'}
   5. Most popular drink: {drink_counts.index[0] if len(drink_counts)>0 else 'None'}
""")

print("\nSaving cleaned data...")
df_clean.to_csv('gym_membership_cleaned.csv', index=False)
print("Saved as: gym_membership_cleaned.csv")

print("\nSample of cleaned data:")
print("="*40)
print(df_clean[['id', 'gender', 'Age', 'abonoment_type', 'visit_per_week',
                'visit_time', 'age_group', 'total_activities']].head())

print("\n" + "="*60)
print("ANALYSIS COMPLETE!")
print("="*60)
```
